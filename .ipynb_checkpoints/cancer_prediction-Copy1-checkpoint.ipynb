## Breast Cancer dataset Analysis and Prediction

```python
# Import the primarily required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

Checked for highly correlated features and consider dropping redundant ones before clustering. PCA handles this issue.

```python
# Clustering using StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
# Apply KMeans clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
df['Cluster'] = kmeans.labels_
```

```python
df
```

```python
# Checking the performance of the model
from sklearn.metrics import silhouette_score
```

```python
sil_score = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score using StandardScaler: {sil_score:.4f}")
```

The silhouette score using the original scaled features was approximately **0.39**. This indicates that the KMeans algorithm was able to find some structure in the data, but the separation between the two clusters is moderate. It suggests potential overlap between cluster boundaries in the original feature space.


```python
# Hyperparameter Tuning with StandardScaler
best_score = -1
best_params = {}

for n_init in [10, 20, 50, 100]:
    for max_iter in [300, 500, 1000]:
        kmeans = KMeans(n_clusters=2, n_init=n_init, max_iter=max_iter, random_state=42)
        labels = kmeans.fit_predict(X_scaled)
        score = silhouette_score(X_scaled, labels)
        
        print(f"n_init={n_init}, max_iter={max_iter} => Silhouette Score: {score:.4f}")
        
        if score > best_score:
            best_score = score
            best_params = {'n_init': n_init, 'max_iter': max_iter}

print("\nBest Parameters:", best_params)
print("Best Silhouette Score:", best_score)
```

```python
# Clustering using the best parameters
kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

sil_score = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score using StandardScaler after Hyperparameter Tuning: {sil_score:.4f}")
```

After experimenting with different values of `n_init` and `max_iter`, the best silhouette score was still around **0.39**, but with improved stability. This shows that while tuning doesn't drastically change performance, it helps the KMeans algorithm converge more reliably by reducing randomness in centroid initialization.


```python
# Performing PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_pca)
```

```python
# Checking the performance with the best parameters after PCA
score = silhouette_score(X_pca, clusters)
print(f"Silhouette Score after PCA: {score:.4f}")
```

Applying PCA with 2 components increased the silhouette score to approximately **0.49**, which is a noticeable improvement. This indicates that dimensionality reduction helped reveal a clearer cluster structure, likely by removing noise and redundant feature information. It also made cluster visualization easier in 2D space.


```python
# Plotting the data
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set1')
plt.title('2-Cluster KMeans on Breast Cancer Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```

```python
# Finding the Best Parameter for Hyperparameter Tuning for PCA
pca = PCA().fit(X_scaled)
explained_var = np.cumsum(pca.explained_variance_ratio_)

plt.plot(range(1, len(explained_var)+1), explained_var, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--')  # 95% variance line
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance')
plt.grid(True)
plt.show()
```

Based on the graph above, at 4 components, the curve crosses the 95% (0.95) red line. This means the first 4 principal components retain at least 95% of the original dataset's variance. So, n_components = 4 and this is the optimal number to use for PCA if your goal is to reduce dimensionality while keeping at least 95% of the information.

```python
# Using n_components=4 for PCA and best parameters for kmeans
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)
kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_pca)

score = silhouette_score(X_pca, clusters)
print(f"Silhouette Score after PCA (with 4 features): {score:.4f}")
```

Using 4 PCA components captured more variance (~95%), but the silhouette score slightly decreased. This suggests that additional components may have reintroduced noise or complexity that blurred the cluster boundaries. Sometimes, a simpler feature space is more effective for clustering, even if it explains slightly less variance.


```python
# Using n_components=3 for PCA and best parameters for kmeans
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_pca)

score = silhouette_score(X_pca, clusters)
print(f"Silhouette Score after PCA (with 3 features): {score:.4f}")
```

Using 3 PCA components captured variance close to 90%, but the silhouette score slightly decreased from what is used as 2 PCA. This also suggests that additional components may have reintroduced noise or complexity that blurred the cluster boundaries. Sometimes, a simpler feature space is more effective for clustering, even if it explains slightly less variance.

```python
# Clustering using zscore
from scipy.stats import zscore
```

```python
z_scores = np.abs(zscore(df))
df_clean = df[(z_scores < 3).all(axis=1)]
```

```python
df_clean
```

```python
df_clean = df_clean.drop(['PCA1', 'PCA2'],  axis=1)
```

```python
df_clean
```

```python
X_scaled = scaler.fit_transform(df_clean)
```

```python
kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
sil_score = silhouette_score(X_scaled, clusters)
print(f"Silhouette Score after dropping the outliers: {sil_score:.4f}")
```

Outliers were identified and removed using the Z-score method to check if they were negatively affecting the clustering performance. However, the silhouette score after removing outliers was **0.5882**, which, while numerically higher, does not represent a significant improvement over previous results. This suggests that the presence of outliers was not a major factor limiting the clustering quality. KMeans inherently handles some variance in data structure, and in this case, the core cluster structure remained largely unchanged with or without outliers.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_pca)


score = silhouette_score(X_pca, clusters)
print(f"Silhouette Score after PCA: {score:.4f}")
```

After removing outliers using the Z-score method and then applying PCA with 2 components, the silhouette score improved substantially to **0.7038**. This indicates a much clearer separation between the two clusters. Removing the outliers likely reduced noise and distortion in the dataset, allowing PCA to project the core structure of the data more effectively. This result demonstrates that **combining outlier removal with dimensionality reduction can significantly enhance clustering performance**, especially in unsupervised tasks like KMeans.


```python
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)

kmeans = kmeans = KMeans(n_clusters=2, n_init=10, max_iter=300, random_state=42)
clusters = kmeans.fit_predict(X_pca)


score = silhouette_score(X_pca, clusters)
print(f"Silhouette Score after PCA: {score:.4f}")
```

### Final Decision Summary:

Based on all evaluations, using **PCA with 2 components** with the **MinMaxScaler** feature scaling method provides the best **Silhouette Score of 0.7182** and the **best trade-off** between performance and interpretability. It delivers the highest silhouette score and enables effective 2D visualization, making it the most suitable setup for this unsupervised clustering task.
